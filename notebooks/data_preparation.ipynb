```python
import pandas as pd
```

```python
features = pd.read_csv('../data/acoustic_features.csv', sep='\t')
features
```

```python
songs = pd.read_csv('../data/songs.csv', sep='\t')
songs
```

```python
artists = pd.read_csv('../data/artists.csv', sep='\t')
artists
```

```python
score = pd.read_csv('../data/song_pop.csv', sep='\t')
score
```

```python
song_id_counts = score.song_id.value_counts()
song_id_counts
```

```python
most_frequent_song_id = song_id_counts.idxmax()
most_frequent_count = song_id_counts.max()

most_frequent_song_data = score[score['song_id'] == most_frequent_song_id]
years = most_frequent_song_data['year'].unique()

print(f"The most frequent song_id is: {most_frequent_song_id}")
print(f"It appears {most_frequent_count} times.")
print(f"It appears in the following years: {years}")
```

```python
score_sorted = score.sort_values(by=['song_id', 'year_end_score'], ascending=[True, False])
score = score_sorted.groupby('song_id', as_index=False).first()[['song_id', 'year_end_score', 'year']]
score = score.rename(columns={'year_end_score': 'score', 'year': 'release_year'})

score
```

```python
score.song_id.nunique()
```

```python
df = pd.merge(score, features, on='song_id', how='left')
df
```

```python
import ast

songs['artist_id'] = songs['artists'].apply(lambda x: list(ast.literal_eval(x).keys())[0])
songs = songs.drop(columns=['artists', 'song_name', 'billboard', 'popularity'])
songs
```

```python
df = pd.merge(df, songs, on='song_id', how='left')
df
```

```python
df = pd.merge(df, artists[['artist_id', 'followers','popularity','artist_type', 'main_genre']], on='artist_id', how='left')
df
```

```python
df = df.drop(columns=['song_id', 'artist_id'])
df
```
